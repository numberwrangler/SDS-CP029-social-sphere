```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Not Affected", "Affected"], loc=legend_)
    # Labels
    plt.ylabel("Student base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distribution in a stacked histogram of affected or not affected students
    """
    # Use the actual column name from your dataset
    temp = pd.DataFrame({
        "Not Affected": dataframe[dataframe["affects_academic_performance"] == 0][column],
        "Affected": dataframe[dataframe["affects_academic_performance"] == 1][column]
    })

    # Plot the histogram
    temp[["Not Affected", "Affected"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    
    # Labels and formatting
    #ax.set_xlabel(column)
    ax.set_ylabel("Count")
    ax.set_title(f"Distribution of {column} by Academic Impact")
    ax.ticklabel_format(style='plain', axis='x')
```

**üß† Insight:**

Based on the distribution of the Conflicts_Over_Social_Media variable, it's evident that this is a multi-class classification task. However, the class frequencies are not evenly distributed, which could lead to performance issues due to class imbalance.

To address this, I'll regroup the classes into three broader categories: Low Conflict, Moderate Conflict, and High Conflict. This grouping strategy is aimed at achieving a more balanced class distribution, which is crucial for building a robust and fair classification model.

0, 1, 2 : Low Conflict

3   : Moderate Conflict

4, 5  : High Conflict

If I were to use binary classification instead, I‚Äôd likely encounter class imbalance ‚Äî since grouping 4 and 5 alone as ‚ÄúHigh Conflict‚Äù would result in very few samples, while 0‚Äì3 would dominate the ‚ÄúLow Conflict‚Äù class. I also thought of  including 3 in the high-conflict group, but that feels like a stretch, as 3 represents moderate behavior, and it still wouldn‚Äôt completely resolve the imbalance.

```python
task2_df.head(2)
```

```python

conflict_mapping = {
    0: 'Low Conflict',
    1: 'Low Conflict',
    2: 'Low Conflict',
    3: 'Moderate Conflict',
    4: 'High Conflict',
    5: 'High Conflict'
}
```

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars for conflict levels with annotations.
    Assumes dataframe contains 3 columns: Low Conflict, Moderate Conflict, High Conflict
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)

    # Update legend to reflect conflict levels
    plt.legend(["Low Conflict", "Moderate Conflict", "High Conflict"], loc=legend_)

    # Axis labels
    plt.ylabel("Student base (%)")
    plt.tight_layout()
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """
    for p in ax.patches:
        value = str(round(p.get_height(), 1))
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x() + p.get_width() / 2) * pad - 0.05, (p.get_y() + p.get_height() / 2) * pad),
            color=colour,
            size=textsize
        )
```

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distribution in a stacked histogram across conflict levels in relationships due to social media.
    """
    # Create a DataFrame for the three conflict levels
    temp = pd.DataFrame({
        "Low Conflict": dataframe[dataframe["conflict_level_in_relationship_over_social_media"] == 0][column],
        "Moderate Conflict": dataframe[dataframe["conflict_level_in_relationship_over_social_media"] == 1][column],
        "High Conflict": dataframe[dataframe["conflict_level_in_relationship_over_social_media"] == 2][column]
    })

    # Plot the stacked histogram
    temp[["Low Conflict", "Moderate Conflict", "High Conflict"]].plot(
        kind='hist', bins=bins_, ax=ax, stacked=True, alpha=0.7
    )

    # Labels and formatting
    ax.set_ylabel("Count")
    ax.set_title(f"Distribution of {column} by Conflict Level")
    ax.ticklabel_format(style='plain', axis='x')
```
